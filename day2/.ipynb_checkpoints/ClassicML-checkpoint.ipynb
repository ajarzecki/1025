# Train/Test Split

scikit-learn provides the tools to easily split data into train/test sets. It's also possible to stratify them based on classes, groups or both, but we don't do it here.

- We keep the target values (MedHouseVal) Serie in `full_y`
- We store all the other columns (our data set with MedHouseVal dropped) in `full_X`

To make sure both full_X and full_y are sampled accordingly, we pass both of them to the `train_test_split` function.

```python
from sklearn.model_selection import train_test_split

full_y = data2["MedHouseVal"]
full_X = data2.drop(columns="MedHouseVal")

train_X, test_X, train_y, test_y = train_test_split(full_X, full_y, train_size=0.8, random_state=42, shuffle=True)
```

# Training

For all trainings we'll use `train_X` as the predictors and `train_y` as the target values. For tests, we'll use `test_X` predictors and their corresponding `test_y`.

We'll start with a simple [linear regression](https://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares) and check the MSE metrics. Then we'll head to more advanced ML algorithms.

## Linear Regression

We train models with `fit`, then call `predict` on the test dataset and calculate MSE.

Using the built-in metrics helpers (`mean_square_error` in our case) makes the code short and clean.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Fitting
lin_reg = LinearRegression()
lin_reg.fit(train_X, train_y)

# Prediction
predicted_y = lin_reg.predict(test_X)

# Metrics
print("MSE", mean_squared_error(test_y, predicted_y))
```

As we'll use the above pattern several times, let's define a helper function not to repeat ourselves.

To better understand how our model performs and generalizes, we'll print both - train set MSE as well as test set MSE.

```python
def fit_and_estimate_model(model):
    model.fit(train_X, train_y)

    print("----- Training -----")
    print("MSE", mean_squared_error(train_y, model.predict(train_X)))

    print("----- Test -----")
    print("MSE", mean_squared_error(test_y, model.predict(test_X)))

fit_and_estimate_model(LinearRegression())
```

We have a comparable MSE for both Test and Training. Model isn't overfitting, however MSE seems large. Let's see what we can do with it.

## Polynomial Regression

A simple linear regression can be extended by constructing [polynomial features](https://scikit-learn.org/stable/modules/linear_model.html#polynomial-regression-extending-linear-models-with-basis-functions) from the coefficients 

In the standard linear regression case, we might have a model that looks like this for two-dimensional data:

$\hat{y}(w, x) = w_0 + w_1 x_1 + w_2 x_2$

If we want to fit a paraboloid to the data instead of a plane, we can combine the features in second-order polynomials, so that the model looks like this:

$\hat{y}(w, x) = w_0 + w_1 x_1 + w_2 x_2 + w_3 x_1 x_2 + w_4 x_1^2 + w_5 x_2^2$

The (sometimes surprising) observation is that this is still a linear model: to see this, imagine creating a new set of features

$z = [x_1, x_2, x_1 x_2, x_1^2, x_2^2]$

With this re-labeling of the data, our problem can be written

$\hat{y}(w, z) = w_0 + w_1 z_1 + w_2 z_2 + w_3 z_3 + w_4 z_4 + w_5 z_5$

We could create the polynomial features manually and then pass them to the LinearRegression model, however it's more convenient to make [pipelines](https://scikit-learn.org/stable/modules/compose.html) in such a case. Pipelines can be called in the same way as the standard estimators with `fit` and `predict` methods.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

model = make_pipeline(
    PolynomialFeatures(degree=2),
    LinearRegression()
)

fit_and_estimate_model(model)
```

Polynomial regression decreased the training MSE, but the Test MSE is much larger. We clearly have a model overfitting the training set.

Let's take a look at the following diagram
https://scikit-learn.org/dev/tutorial/machine_learning_map/index.html to decide what we should check next.

## Linear Regression with Lasso Regularization

Regularization is introduced to prevent overfitting. It does this by adding a penalty term to the loss function that the algorithm tries to minimize. This penalty is based on the size of the coefficients.

1. [Lasso](https://scikit-learn.org/stable/modules/linear_model.html#lasso) (L1 Regularization) penalizes the absolute values of the coefficients. It can force some coefficients to be exactly zero, leading to feature selection (by clearing some of the variables).

2. [Ridge](https://scikit-learn.org/stable/modules/linear_model.html#ridge-regression-and-classification) (L2 Regularization) penalizes the square of the coefficients. It tends to shrink coefficients uniformly but usually doesn't force them to zero.

3. [ElasticNet](https://scikit-learn.org/stable/modules/linear_model.html#elastic-net) is a hybrid of ridge and lasso regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge, Lasso, ElasticNet

model = make_pipeline(
    PolynomialFeatures(degree=2),
    Lasso(random_state=42, alpha=.25)
)

fit_and_estimate_model(model)
```

OK, looks a little better, but let's continue the exploration of other methods to see if we can improve the results further.

## Ensembles

[Ensemble methods](https://scikit-learn.org/stable/modules/ensemble.html#gradient-boosted-trees) combine the predictions of several base estimators built with a given learning algorithm in order to improve generalizability / robustness over a single estimator.

Two very famous examples of ensemble methods are gradient-boosted trees and random forests.

### Gradient Boosting Regressor

Scikit-learn provides two implementations of gradient-boosted trees: HistGradientBoosting and GradientBoosting.

The former can be orders of magnitude faster than the latter when the number of samples is larger than tens of thousands of samples.

Missing values and categorical data are natively supported by the Hist… version, removing the need for additional preprocessing such as imputation.

```python
import numpy as np
from sklearn.model_selection import validation_curve, learning_curve
from sklearn.ensemble import HistGradientBoostingRegressor

fit_and_estimate_model(HistGradientBoostingRegressor(random_state=42))
```

Much better now ;) but we can still tweak some hiperparameters. As we don't have a separate validation set, we'll use k-fold cross-validation.

### Cross Validation

In k-fold cross validation:
1. Training data is split into k parts.
1. Model is trained on k-1 of the folds.
1. The remaining 1 fold is used as the validation set.
1. Entire operation is repeated so that it's performed k times in total.

`cross_val_score` allows us to perform k-fold cross validation (default is 5 folds, but can be changed with `cv` parameter) and returns an array of k scores (one per each fold).

This allows us to use entire training set and still estimate its performance on previously unseen data, without "leaking" test dataset into the process. 

```python
from sklearn.model_selection import cross_val_score

-cross_val_score(
    HistGradientBoostingRegressor(random_state=42), 
    train_X, 
    train_y, 
    scoring='neg_mean_squared_error').mean()
```

We'll use cross-validation during hyperparameters tuning. 

Keep in mind the training process will be ~ k times slower, as the model needs to be trained that many times.

### Grid Search

With `GridSearchCV` we can define the parameter space (values we want to validate).

`fit` creates all possible candidate sets for parameters, trains the model as many times as needed and uses cross-validation for each set.

Because of this, it's useful to parallelize the proces using `n_jobs` (-1 means use all cores).

Scoring functions can be found [under this link](https://scikit-learn.org/dev/modules/model_evaluation.html#scoring-parameter)

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
  {
      'l2_regularization': [0, 0.1, 0.5, 1], 
      'max_leaf_nodes': [10, 20, 31],
      'max_depth': [3, 5, 10, 20, None],
      'min_samples_leaf': [5, 20, 30]
  },
]

estimator = HistGradientBoostingRegressor(random_state=42)

search = GridSearchCV(
    estimator, 
    param_grid, 
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)
search.fit(train_X, train_y)
```

Let's see what values perform best. 

```python
print(search.best_params_)
```

Now we can train the model with these parameters on full training set (without cross-validating it) and use the test dataset for the final results

```python
fit_and_estimate_model(
    HistGradientBoostingRegressor(
        random_state=42, 
        l2_regularization = 0.1,
        max_depth=20,
        max_leaf_nodes=31, 
        min_samples_leaf = 20
    )
)
```

### Randomized Search

RandomizedSearchCV is another (random) way to find the best hyperparameters. 

This time, instead of proposed values to check, we describe the distributions (via `scipy.stats` for instance).

Because there are no upfront combinations of params, we define how many tests we want to perform (with `n_iter`).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import SGDRegressor
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV

param_distributions = {
    'l2_regularization': uniform(0, 1),
    'max_depth': randint(5, 30),
    'max_leaf_nodes': randint(5, 30),
    'min_samples_leaf': randint(5, 30)
}

estimator = HistGradientBoostingRegressor(random_state=42)

rnd_search = RandomizedSearchCV(
    estimator, 
    param_distributions=param_distributions, 
    n_iter=300,
    cv=5, 
    scoring='neg_mean_squared_error', 
    random_state=42,
    n_jobs=-1
)

rnd_search.fit(train_X, train_y)
# fit_and_estimate_model(rnd_search)
```

```python
rnd_search.best_params_
```

```python
fit_and_estimate_model(HistGradientBoostingRegressor(random_state=42, l2_regularization=0.8, max_leaf_nodes=29, max_depth=15, min_samples_leaf=29))
```

Looks like our GridSearch led us to slightly better results

# Stochastic Gradient Descent

SGD is a universal training method and `SGDRegressor` uses SGD to train Linear model. 

Based on the above experiments, we don't expect a linear model to perform better than Ensembles.

The only reason we have SGD in this notebook is to show how to draw Train/Validation Curves :)

```python
from sklearn.linear_model import SGDRegressor

sgd = SGDRegressor(loss='squared_error', random_state=42)
sgd.fit(train_X, train_y)

mean_squared_error(test_y, sgd.predict(test_X))
```

Standard SGD yields horrible results! Why? Because it expects features to be scaled. 

Also the default learning rate in our case is way too large, so let's decrease it (via eta0 parameter)

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import make_pipeline

estimator = make_pipeline(
    StandardScaler(),
    SGDRegressor(
        loss='squared_error', 
        eta0=0.001, 
        random_state=42
    )
)
estimator.fit(train_X, train_y)

mean_squared_error(test_y, estimator.predict(test_X))
```

# Train / Validation Curves

The logic here is quite simple:

1. we set `max_iter` to 1 so that calling `fit` will perform stop after one epoch
1. we set `warm_start` to `true` because otherwise `fit` would start from the beginning with consecutive calls
1. after each round we calculate train- and test- MSE and add the results to two arrays
1. we plot the arrays when we finish

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

estimator = make_pipeline(
    StandardScaler(),
    SGDRegressor(
        loss='squared_error', 
        eta0=0.001, 
        random_state=42,
        warm_start=True,
        max_iter=1
    )
)

train_loss = []
test_loss = []

for _ in range(100):
    estimator.fit(train_X, train_y)
    train_loss.append(mean_squared_error(train_y, estimator.predict(train_X)))
    test_loss.append(mean_squared_error(test_y, estimator.predict(test_X)))

plt.plot(train_loss)
plt.plot(test_loss)
```

# Validation Curve

`ValidationCurveDisplay` allows visualizing cross-validation for a single parameter. 

Let's create a pipeline, check what parameters it has and validate several settings for one of them

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

estimator = make_pipeline(
    StandardScaler(),
    SGDRegressor(
        random_state=42,
    )
)
estimator.get_params()
```

Results contain an array of arrays (one for each provided value) with k-fold validation scores

```python
from sklearn.model_selection import ValidationCurveDisplay

param_range = np.linspace(0.001, 0.02)

ValidationCurveDisplay.from_estimator(
    estimator, 
    train_X, 
    train_y, 
    param_name='sgdregressor__eta0', 
    param_range=param_range,
    scoring='neg_root_mean_squared_error'
)
```
