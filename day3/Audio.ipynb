# Automatic Speech Recognition

Keep in mind, Gradio asynchronously saves file. Don't hit submit button until your wave is available and visible inside gradio

```python
from transformers import pipeline

transcriber = pipeline(task="automatic-speech-recognition", model="openai/whisper-base")

import gradio as gr

def recognize(audio_path):
    return transcriber(audio_path, generate_kwargs = {"task":"transcribe", "language":"<|pl|>"} )

gr.Interface(
    fn=recognize,
    inputs=[gr.Audio(type="filepath")], # you can change this to gr.Microphone()
    outputs=["text"]
).launch()
```
